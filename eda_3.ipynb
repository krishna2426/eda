```python
train_df = pd.read_excel('Data_Train.xlsx')
train_df.head()
```

```python
test_df = pd.read_excel('Test_set.xlsx')
test_df.head()
```

```python
final_df = pd.concat([train_df, test_df])
final_df.head()
```

```python
final_df.info()
```

```python
# creating derived features from Date_of_Journey field
final_df['Date'] = final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month'] = final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year'] = final_df['Date_of_Journey'].str.split('/').str[2]
final_df.head()

#MY solution
#final_df['Date_of_Journey'] = pd.to_datetime(final_df['Date_of_Journey'], format='%d/%m/%Y')
#final_df['Date'] = final_df['Date_of_Journey'].dt.day
#final_df['Month'] = final_df['Date_of_Journey'].dt.month
#final_df['Year'] = final_df['Date_of_Journey'].dt.year
#final_df.head()
```

```python
# convert date, month and year into integr
final_df['Date'] = final_df['Date'].astype(int)
final_df['Month'] = final_df['Month'].astype(int)
final_df['Year'] = final_df['Year'].astype(int)
final_df.info()
```

```python
final_df.drop('Date_of_Journey', axis=1, inplace=True)
final_df.head()
```

```python
# arrival time splitting for time only
final_df['Arrival_Time'] = final_df['Arrival_Time'].str.split(' ').str[0]
final_df.head()
```

```python
final_df.isnull().sum()
```

```python
# taking out hours and minutes from arrival time
final_df['Arrival_hour'] = final_df['Arrival_Time'].str.split(':').str[0]
final_df['Arrival_min'] = final_df['Arrival_Time'].str.split(':').str[1]
final_df.head()
```

```python
# obj --> integer
final_df['Arrival_hour'] = final_df['Arrival_hour'].astype(int)
final_df['Arrival_min'] = final_df['Arrival_min'].astype(int)
final_df.info()
```

```python
final_df.drop('Arrival_Time', axis=1, inplace=True)
final_df.head()
```

```python
final_df['Dep_hour'] = final_df['Dep_Time'].str.split(':').str[0]
final_df['Dep_min'] = final_df['Dep_Time'].str.split(':').str[1]
final_df.head()
```

```python
# obj --> int
final_df['Dep_hour'] = final_df['Dep_hour'].astype(int)
final_df['Dep_min'] = final_df['Dep_min'].astype(int)
final_df.drop('Dep_Time', axis=1, inplace=True)
final_df.info()
```

```python
# route --> caturing every route
# we can drop route, and focus on total stops ,
final_df['Total_Stops'].unique()
# map the values
final_df['Total_Stops'] = final_df['Total_Stops'].map({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3, 'nan':1, '4 stops':4})
final_df.head()
```

```python
#drop route
final_df.drop('Route',axis=1, inplace=True)
final_df.head()
```

```python
# duration into min
final_df['duration_hour']=final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
final_df.head()
```
