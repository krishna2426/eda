```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns #we see that the country columns exist in both the datasets ,we will try to combine them
```

```python
final_df = pd.merge(df, df_country, on='Country Code', how='left')   ## 'on' parameter for waht we want to join(specific column) , 'how' parameter on what type of join it will be
```

```python
## to check datatypes another way
final_df.dtypes
```

```python
final_df.columns
```

```python

country_names = final_df.Country.value_counts().index #trying to find out how many countries are there, and with a specific country how many records are there
country_names
```

```python
country_vlaues = final_df.Country.value_counts().values #just the values
```

# ***Observations : Zomato maximum records and transactions are from india followed by US and the UK***

```python
#pie chart -top  3 countries that uses zomato
plt.pie(country_vlaues[:3], labels=country_names[:3], autopct='%1.2f%%')
```

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

##obs:
1. when rating was 4.5 to 4.9 --> excellent
2. when rating 4.0 to 4.5 --> very good
3. when rating 3.5 to 3.9 --> good
4. when rating 2.5 to 3.4 --> average
5. when rating 1.8 to 2.4 --> poor


```python
 ratings.head()
```

```python

plt.rcParams['figure.figsize'] = (12,6)
sns.barplot(x= 'Aggregate rating', y='Rating Count', data=ratings)
```

```python
 sns.barplot(x= 'Aggregate rating', y='Rating Count', hue='Rating color',data=ratings, palette=['white', 'red', 'orange', 'yellow', 'green', 'green'])
```

Observations:
1. Not rated Count is very high
2. Maximum Number of rating is between 2.5 to 3.4


```python
##count plot is used for categorical value
sns.countplot(x='Rating color', data=ratings, palette=['white', 'red', 'orange', 'yellow', 'green', 'green'])
```
